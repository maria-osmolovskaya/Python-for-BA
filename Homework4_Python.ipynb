##**Условие 2: 2 задача**

## 2.1. Изучите распределение домов от наличия вида на набережную


```python
def check_waterfront(x):
  if x == 0:
    return 'Нет вида'
  else:
    return 'Есть вид'

data['waterfront'] = data['waterfront'].apply(check_waterfront)
data.head()
```

```python
data2 = data['waterfront'].value_counts()
data2.head()
```

```python
plt.figure(figsize=(6,4))
plt.pie(data2, autopct = '%1.1f%%')
plt.title('Наличие вида на набережную')
plt.legend(data2.index)
```

## 2.2. Изучите распределение этажей домов

```python
data3 = data['floors'].value_counts()
data3.head()
```

```python
plt.figure(figsize=(6,4))
plt.bar(data3.index, data3.values)
plt.title('Структура недвижимости по количеству этажей')
plt.xlabel('Количество этажей')
plt.ylabel('Количество объектов')
```

##**Условие 3: 3 задача**

## Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
sns.jointplot(x = data['grade'], y = data['price'], kind = 'reg')
plt.title('Зависимость стоимости объекта недвижимости от состояния', pad = 80)
```

C повышением уровня состояния растет стоимость недвижимости, при этом - чем выше состояние, тем больше разброс стоимости объектов.

```python
sns.jointplot(x = data['sqft_living'], y = data['price'], kind = 'reg')
plt.title('Зависимость стоимости объекта недвижимости от жилой площади', pad = 80)
```

Наблюдается прямая корреляция между жилой площадью и стоимостью объектов недвижимости.

```python
plt.figure(figsize=(8,2))
sns.boxplot(x = data['price'], y = data['view'], whis = 15)

plt.title('Зависимость стоимости объекта недвижимости от наличия вида')
plt.xlabel('price')
plt.ylabel('view')
```
