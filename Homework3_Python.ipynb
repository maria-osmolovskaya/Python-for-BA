## **Задание 1**
### Считать данные с помощью pandas. Вывести на экран первые 5 строк

```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv', sep = ',')
df.head()
```

### 1.1 Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. метр общей площади

```python
df['price_per_sq_lot'] = df['price'] / df['sqft_lot']
df.head()
```

### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома Если реновации дома не было, то в новом признаке поставьте 0.

```python
df['delta_renovated'] = df['yr_renovated'] - df['yr_built']

def check_renovation(x):
  if x < 0:
    return 0
  else:
    return x

df['delta_renovated'] = df['delta_renovated'].apply(check_renovation)
df.head()
```

## 1.3 Создайте признаки года продажи, месяца продажи

```python
df['yr_sold'] = 0

def year(x):
  x = x[0:4]
  return x

df['yr_sold'] = df['date'].apply(year)
df.head()
```

```python
df['month_sold'] = 0

def month(x):
  x = x[4:6]
  return x

df['month_sold'] = df['date'].apply(month)
df.head()
```

## 1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns = ['date', 'zipcode', 'lat', 'long'], inplace = True)
df.head()
```
